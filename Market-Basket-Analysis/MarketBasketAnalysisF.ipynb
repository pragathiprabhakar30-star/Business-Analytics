```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

##  Dataset Description

The dataset used in this project is a grocery transactions dataset obtained from Kaggle. It contains records of items purchased by customers over different transactions.

###  Dataset Features:
- **Member_number**: Unique identifier for each customer
- **Date**: Date of transaction
- **itemDescription**: Name of the purchased product

Each row represents a single item purchased in a transaction. Multiple rows with the same Member ID and Date represent items bought together.

```python
data = pd.read_csv('/content/Groceries_dataset1.csv')
data.head()
```

```python
print(data.columns)
```

```python
data.columns = ['Member_number', 'Date', 'itemDescription']
```

```python
transactions = data.groupby(['Member_number', 'Date'])['itemDescription'].apply(list)
transactions = transactions.tolist()

transactions[:5]
```

```python
te = TransactionEncoder()
te_data = te.fit(transactions).transform(transactions)

df = pd.DataFrame(te_data, columns=te.columns_)
df.head()
```

## Frequent Itemset Generation (Apriori Algorithm)

The Apriori algorithm is applied to identify frequently occurring item combinations in the dataset. It works by:

- Identifying individual items with minimum support
- Expanding them into larger itemsets
- Pruning itemsets that do not meet the support threshold

Minimum support is used to control how frequently an itemset must appear to be considered significant.

```python
frequent_itemsets = apriori(df, min_support=0.003, use_colnames=True)
frequent_itemsets.head()
```

##  Association Rule Mining

Association rules are generated from the frequent itemsets to identify relationships between products.

Each rule is evaluated using:

- **Support**: Frequency of occurrence of an itemset
- **Confidence**: Likelihood of purchasing consequent given antecedent
- **Lift**: Strength of the relationship between items

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.05)

rules = rules.sort_values(by='lift', ascending=False)
rules.head()
```

```python
strong_rules = rules
strong_rules.head()
```

```python
strong_rules['antecedents'] = strong_rules['antecedents'].apply(lambda x: ', '.join(list(x)))
strong_rules['consequents'] = strong_rules['consequents'].apply(lambda x: ', '.join(list(x)))

strong_rules.to_csv('association_rules_clean.csv', index=False)
```

```python
print("Frequent Itemsets:", frequent_itemsets.shape)
print("Rules:", rules.shape)
print("Strong Rules:", strong_rules.shape)
```

## Exporting Results

The generated association rules are cleaned and exported into a CSV file. This file is later used in Power BI for visualization and dashboard creation.

The antecedents and consequents are converted into readable text format for better understanding and analysis.

```python
strong_rules.to_csv('association_rules.csv', index=False)
```
